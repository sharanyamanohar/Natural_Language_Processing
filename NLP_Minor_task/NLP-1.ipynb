### Listing out capital letters from text column

```python
df=pd.read_excel(r"Capital_Voice Data_Disc_1_WITH ROW NUMBERS.xlsx")
df.head()
```

```python
df['Text']
```

```python
import re

capital_words = []#Listing out all the capital words

for text in df['Text']:
    # Find all capital words in the text using regular expressions
    words = re.findall(r'\b[A-Z]+\b', text)
    capital_words.extend(words)#Same as append

print(capital_words)
```

### Extracting the rows of Capital Letters words 

```python
word_rows = []  

for index, text in enumerate(df['Text']):#For holding the index values we are doing with enumerate
    word_rows.extend([index] * len(words))  # Repeat the row index for each capital word found
    # word_rows contains the list of all rows(It tracks out the rows in which capital letter is found)
for word,row in zip(capital_words, word_rows):
    print(f"Word: {word}, Row: {row}")
```
