### NLP-Convert letter into lower case for all non medical terms 

```python
import pandas as pd
df=pd.read_excel("Capital_ClinMed_part 8-10 hours_revised_Swapnil_WITH ROW NUMBERS.xlsx")
```

```python
df['Text']
```

```python
file_path = 'extract.txt'  

with open(file_path, 'r') as file:
    content = file.read()

words = content.split()

print(words)
```

```python
file_path = 'get.txt'  

with open(file_path, 'r') as file:
    content = file.read()
extract = content.split()

print(extract)
```

```python
import re

capital_words = []  # Listing out all the capital words

for text in df['Text']:
    # Find all capital words in the text using regular expressions
    words = re.findall(r'\b[A-Z]+\b', text)
    capital_words.extend(words)  # Same as append

updated_text = []

for word in capital_words:
    if word in extract:
        updated_text.append(word.lower())  # Convert matching capital words to lowercase
    else:
        updated_text.append(word)  # Keep non-matching capital words as it is

print(updated_text)
```
